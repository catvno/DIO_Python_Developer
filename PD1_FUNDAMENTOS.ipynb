## Tipos de Operadores com Python

### 1. Operadores Aritméticos

Os operadores aritméticos são responsáveis por implementar as operações matemáticas>

```python
# Adição:
print(1 + 3)

# Subtração:
print(12 - 4)

# Multiplicação:
print(6 * 7)

# Divisão:
print(21 / 3)

# Resto:
print(10 % 3)

# Exponenciação:
print(2 ** 3)

# Divisão Inteira:
print(21 // 2)
```

Na matemática existe uma regra que indica quais operações devem ser executadas primeiro. Isto é útil pois ao analisar uma expressão, a depender da ordem das operações, o resultado pode ser diferente.

Um exemplo é: x = 10 - 5 * 2

O resultado é 10 ou 0?

```python
print(10 - 5 * 2)
```

Veja que, assim como na matemática, a máquina respeira a "hierarquia" das operações. A hierarquia utilizada é:

* Parênteses;
* Expoentes;
* Multiplicação e divisões (esquerda para direita);
* Somas e subtrações (esquerda para direita).

```python
print((10 - 5) * 2)
```

### 2. Operadores de Comparação

Operadores de comparação são utilizados para comparar suas variáveis com valores. Podem ser utilizados como condições ao longo do código.

```python
saldo = 450
saque = 200

# Igualdade:
print(saldo == saque)

# Desigualdade:
print(saldo != saque)

# Maior que:
print(saldo > saque)

# Menor que:
print(saldo < saque)

# Maior ou igual a:
print(saldo >= saque)

# Menor ou igual a:
print(saldo <= saque)
```

Veja que  retorno desse tipo de operador é sempre um booleano (True or False).

### 3. Operadores de Atribuição

Operadores de atribuição são utilizados para definir o valor inicial ou sobreescrever o valor de uma variável.

```python
# Atribuição simples:
saldo  = 500
print(saldo)

# Atribuição com soma:
saldo += 200
print(saldo)

# Atribuição com subtração:
saldo -= 300
print(saldo)

# Atribuição com multiplicação:
saldo *= 2
print(saldo)

# Atribuição com divisão:
saldo /= 2
print(saldo)

# Atribuição com exponenciação:
saldo **= 2
print(saldo)

# Atribuição com divisão inteira:
saldo //= 2
print(saldo)

# Atribuição com resto:
saldo %= 3
print(saldo)
```

### 4. Operadores Lógicos

Os operadores lógicos são utilizados em conjunto com coperadores de comparação, para montar expressões lógicas. Quando um operador de comparação é utilizado, o resultado é retornado em booleano, dessa forma podemos combiná-los com operadores lógicos.

```python
# E/and:
print(True and True)
print(True and False)
print(False and True)
print(False and False)

# Ou/or:
print(True or True)
print(True or False)
print(False or True)
print(False or False)

# Não/not:
print(not True)
print(not False)

```

```python
saldo = 1000
saque = 200
limite = 100

# E/and:
print(saldo > saque and saldo > limite) # todas as condições devem ser verdadeiras

# Ou/or:
print(saldo > saque or saldo < limite ) # pelomenos uma das condições deve ser verdadeira

# Não/not:
print(not saldo > saque) # imprime o booleano oposto
```

Para concatenar operadores lógicos, a hierarquia considerada é da esquerda para direita (ordem natural de leitura). Caso seja desejado alterar a hierarquia, podemos utilizar parênteses.

```python
saldo = 1000
saque = 250
limite = 200
conta_especial = True

saldo >= saque and saldo <= limite or conta_especial and saldo >= saque
```

## Estruturas condicionais e repetições

### 1. Identação e blocos

Identar código é uma forma de manter o código fonte mais legível e manutenível (estética). Mas em python ela exerce um segundo papel, através da identação o interpretador consegue determinar onde um bloco de comando inicia e termina.

```python
def sacar(saldo, valor):
    if saldo >= valor:
        saldo -= valor
        print(f'Saque realizado. Novo saldo atual: {saldo}')
        print('Obrigado por utilizar nosso banco.')
    else:
        print('Saldo insuficiente.')
        print('Obrigado por utilizar nosso banco.')

sacar(300,50)
```

### 2. Estruturas condicionais

A estrutura condicional permite o desvio de fluxo de controle, quando determinadas expressões lógicas são atendidas.

*   **If:** Para criar uma estrutura condicional simples, composta por um único desvio, podemos utilizar a palavra reservada if. O comando irá testar a expressão lógica e em caso de retorno verdadeiro as ações presentes no bloco de código serão executadas.



```python
# Exemplo If

saldo = 2000
saque = float(input("Informe o valor do saque:"))

if saldo >= saque:
    print(f'Saque realizado. Novo saldo atual: {saldo - saque}')
    print('Obrigado por utilizar nosso banco.')

if saldo < saque:
    print('Saldo insuficiente.')
    print('Obrigado por utilizar nosso banco.')
```

* **If/Else:** Para criar uma estrutura condicional com dois desvios, podemos utilizar as palavras reservadas if e else. Como sabemos se a expressão lógica testada no if for verdadadeira, então o bloco de código do if será executado. Caso contrário o bloco de código do else será executado.

```python
# Exemplo If/Else

saldo = 2000
saque = float(input("Informe o valor do saque:"))

if saldo >= saque:
    print(f'Saque realizado. Novo saldo atual: {saldo - saque}')
    print('Obrigado por utilizar nosso banco.')
else:
    print('Saldo insuficiente.')
    print('Obrigado por utilizar nosso banco.')
```

* **If/Elif/Else:** Em alguns cenários queremos mais de dois desvios, para isso podemos utilizar a palavra reservada elif. O elif é composto por uma nova expressão lógica, que será testada e caso retorne verdadeiro o seu bloco de código será executado. Não existe um npumero máximo de elifs que podemos utilizar, porém evite criar grandes estruturas condicionais para não aumentar demasiadamente a commplexidade do código.

```python
# Exemplo If/Elif/Else

opcao = int(input("Informe uma opção: \n [1] Sacar \n [2] Depositar \n [3] Ver extrato \n"))

if opcao == 1:
    print('Você escolheu sacar')
elif opcao == 2:
    print('Você escolheu depositar')
elif opcao == 3:
    print('Você escolheu ver extrato')
else:
    print('Opção inválida')
```

Também podemos utilizar if's aninhados de forma há criar estruturas mais complexas. Para isso, basta adicionar estruturas if/elif/else dentro do bloco de código e de estruturas if/elif/else.

```python
# Estruturas aninhadas

saldo = 2000
cheque_especial = 500
saque = float(input("Informe o valor do saque:"))
conta = int(input("Informe o tipo da conta: \n [1] Normal \n [2] Universitária"))

if conta == 1:
    if saldo >= saque:
        print(f'Saque realizado. Novo saldo atual: {saldo - saque}')
        print('Obrigado por utilizar nosso banco.')
    elif saque <= (saldo + cheque_especial):
        print(f'Saque realizado com cheque especial. Novo saldo atual: {saldo - saque}')
        print('Obrigado por utilizar nosso banco.')
    else:
        print('Saldo insuficiente.')
elif conta == 2:
    if saldo >= saque:
        print(f'Saque realizado. Novo saldo atual: {saldo - saque}')
        print('Obrigado por utilizar nosso banco.')
    else:
        print('Saldo insuficiente.')
else:
    print('Opção inválida')
```

Por fim, temos a condição de if ternário, que permite escrever uma condição em uma única linha. Ele é composto por 3 partes:

* 1: Retorto para expressão verdadeira
* 2: Expressão Lógica
* 3: Retorto para expressão falsa

```python
# If Ternário

saldo = 2000
saque = float(input("Informe o valor do saque:"))

print('Saque realizado' if saldo >= saque else 'Saldo insuficiente')
```

Existem 2 tipos de estruturas de repetição: for e while. A princípio, trabalharemos com o for.

O comando **for** é usado para percorrer um objeto iterável. Faz sentido usar o for quando sabemos o número exato de vezes que nosso bloco de código deve ser executado, ou quando queremos percorrer um objeto iterável.

```python
texto = input("Informe o texto:") # veja que o texto é uma estrutura iterável
VOGAIS = "AEIOU"

for letra in texto:
    if letra.upper() in VOGAIS:
        print(letra, end=' ')
print()
```

O comando **while** é usado para repetir u bloco de código várias vezes. Faz sentido usar o while quando não sabemos o número exato de vezes que nosso bloco de código deve ser executado.

```python
i=0
while i < 30:
  print('Hoje é o dia', i, 'do mês de novembro')
  i += 1
```

### 3. Fatiamento de string

Fatiamento de strings é uma técnica utilizada para retornar substrings (partes da string original), informando o início, o fim e o passo **[starte:stop[,step]]**.

```python
nome = 'Catarina Victória Nascimento de Oliveira'

print(nome[:]+'\n')
print(nome[0]+'\n')
print(nome[:10]+'\n')
print(nome[10:]+'\n')
print(nome[10:20]+'\n')
print(nome[10:20:2]+'\n')
print(nome[::2]+'\n')
print(nome[::-1]+'\n')
```

## Desafio de Código

```python
menu = """
[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair
"""

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:
  opcao = input(menu)

  if opcao == 'd':
    valor = float(input('Informe o valor que deseja depositar: '))
    if valor > 0:
      saldo += valor
      extrato += f'Depósito: R$ {valor:.2f}\n'
      print('Depósito realizado com sucesso.')
    else:
      print('Valor inválido')

  elif opcao == 's':
    valor = float(input('Informe o valor que deseja sacar: '))
    if numero_saques <= LIMITE_SAQUES:
      if valor <= saldo and valor <= limite:
        saldo -= valor
        extrato += f'Saque: R$ {valor:.2f}\n'
        numero_saques += 1
        print('Saque realizado com sucesso.')
      elif valor > limite:
        print(f'O valor informado excede o limite de R${limite}.00 por saque.')
      else:
        print('Saldo insuficiente.')
    else:
      print(f'Você ja atingiu o limite de saques diário correspondente a {LIMITE_SAQUES}.')

  elif opcao == 'e':
    if extrato == "":
      print('Não foram realizadas movimentações.')
    else:
      print('\n================ EXTRATO ================')
      print(f'\nSaldo Atual: R$ {saldo:.2f}')
      print('------------------------------------------')
      print('Histórico de Transações:')
      print(extrato)
      print('==========================================')

  elif opcao == 'q':
    break

  else:
    print('Operação inválida.')
```
